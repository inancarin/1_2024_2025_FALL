# Example 1

```python
x_values = []
for i in range(1, 6):
    x_values.append(i)
for k in range(1, 5):
    #k = 1
    y_values = []
    for val in x_values:
        y_values.append(val * k)

    plt.plot(x_values, y_values, label="y=" + str(k) + "x")
plt.legend()
plt.show()
```
